Purpose is to show how correlation is interepreted with values on x and y axis. 
The purpose of a Pearson correlation test is to measure the strength and direction of the linear relationship between two continuous variables.

0.70 to 1.00 (or -0.70 to -1.00) indicates a strong positive (or negative) linear relationship.
0.50 to 0.70 (or -0.50 to -0.70) indicates a moderate positive (or negative) linear relationship.
0.30 to 0.50 (or -0.30 to -0.50) indicates a weak positive (or negative) linear relationship.
0.00 to 0.30 (or -0.00 to -0.30) indicates a very weak to negligible linear relationship.


Covariance measures the degree to which two variables change together. If the variables tend to increase together, the covariance is positive; if one increases while the other decreases, the covariance is negative. Covariance provides a directional relationship but not the strength of the relationship, unlike correlation.

```python
# calculate the Pearson's correlation between two variables
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr
# seed random number generator
seed(1)

# prepare data
data1 = 10 * randn(100) + 50
data2 = data1 + (10 * randn(100) + 50)

# calculate covariance matrix
covariance = cov(data1, data2)

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)

# plot
plt.scatter(data1, data2)
plt.show()

# summarize
print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearsons correlation: %.3f' % corr)
```
